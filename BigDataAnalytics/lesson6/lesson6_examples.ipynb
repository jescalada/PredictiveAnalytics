```python
from sklearn.linear_model    import LogisticRegression
from sklearn                 import datasets
from sklearn.model_selection import train_test_split
import pandas as pd

# Load data from sklearn library.
df = pd.read_csv("iris_v2.csv")
X = df.copy()
del X['iris_type']
y = df['iris_type']

# Split data.
X_train,X_test,y_train,y_test = train_test_split(
        X, y, test_size=0.25)

# Create one-vs-rest logistic regression object
clf = LogisticRegression(
    random_state=0,
    multi_class='multinomial', solver='newton-cg')

# Train model
model  = clf.fit(X_train, y_train)

# Predict class
y_pred = model.predict(X_test)
print(y_pred)

# Show confusion matrix and accuracy scores.
from   sklearn                 import metrics
cm = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
print('\nAccuracy: ',metrics.accuracy_score(y_test, y_pred))
print("\nConfusion Matrix")
print(cm)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from IPython.display import HTML
```

```python
train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
test_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')
print(train_data.head(3).T) # Transpose so it fits nicely in browser window.
predictor = TabularPredictor(label='class',eval_metric='recall', path='ag-default').fit(train_data, time_limit=120)  # Fit models for 120s
```

```python
leaderboard = predictor.leaderboard(test_data)
print(leaderboard.T)
```
